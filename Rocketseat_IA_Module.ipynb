## Teorema do Limite Central



> Quando você extrai várias amostras aleatórias de uma população e calcula e medida de cada um, independente da forma de distribuição geral, essas medidas se aproximam de uma distribuição normal (Formato de sino) à medida que o tamanho aumenta. nos permitindo fazer inferência de uma população com base em uma amostra



## Colocando em Prática - Introdução a python e estatística

```python
import pandas as pd
print(pd.__version__)
```

```python
dict_medidas = {'idade': [15,18, 25, 25, 40, 55, 58, 60, 80], 'altura': [160, 162, 165, 168, 172, 174, 174, 174, 176]}
```

```python
dict_medidas
```

```python
df_medidas = pd.DataFrame.from_dict(dict_medidas)
```

```python
df_medidas
```

## Medidas Resumo

### Medidas Posição

```python
df_medidas.mean()
```

```python
df_medidas.median()
```

```python
df_medidas.mode()
```

### Medidas de Disperção

```python
# Variancia

df_medidas.var()
```

```python
# Desvio Padrão

df_medidas.std()
```

```python
# Coeficiente de variação

df_medidas.std() / df_medidas.mean() * 100
```

### Medida de Forma

```python
# Assimetria

df_medidas.skew()
```

```python
# Curtose

df_medidas.kurtosis()
```

```python
# Obter medidas estátistica

df_medidas.describe()
```

## Correlações

```python
df_medidas.corr(method="pearson")
```

```python
df_medidas.corr(method="spearman")
```

```python
df_medidas.corr(method="kendall")
```

```python
df_medidas.altura.corr(df_medidas.idade, method="pearson")
```

## Unificando DataFrames

```python
# Quantidade de registro por DataFrame
len(df_customers)
len(df_contracts)
len(df_services)
```

```python
df_services.rename(columns={"customerID": "IDCliente"}, inplace=True)
```

```python
df_services.info()
```

```python
# Unificando DataFrames de Customer e Services

df_temp = df_customers.merge(df_services, on=["IDCliente"])
```

```python
df_temp.info()
```

```python
# Unificar DataFrame Temp com Contract

df_churn_temp = df_temp.merge(df_contracts, left_on=["IDCliente"], right_on=["customerID"])
```

```python
df_churn_temp.info()
```

```python
df_churn = df_customers.merge(df_services, on=["IDCliente"]).merge(df_contracts, left_on=["IDCliente"], right_on=["customerID"])
```

```python
df_churn.info()
```

```python
# Remover Col customerID

df_churn.drop(["customerID"], axis=1, inplace=True)
#  axis=1 determinar que vamos usar coluna
```

```python
df_churn.info()
```

## Imputar Dados nos Valores Ausentes

```python
# Preencher todos os valores ausentes com 0

df_churn.fillna(0)
```

```python
# Preencher valores padrão conforme a coluna

df_churn.fillna(value={"TotalCharges": 0})
```

```python
# Preencher todos os valores ausentes de uma coluna com a media

media_TotalCharges = df_churn.TotalCharges.mean()
media_TotalCharges
```

```python
df_churn.fillna(value={"TotalCharges": media_TotalCharges})
```
